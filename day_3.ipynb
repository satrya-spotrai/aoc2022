```python
with open('./data/day_3.txt') as handle:
    lines = handle.readlines()
    bags = [line[:-1] for line in lines]
```

```python
import string
from collections import defaultdict
```

```python
# priority
priomap = defaultdict(int)
for val, letter in enumerate(string.ascii_letters):
    priomap[letter] = val + 1
```

```python
def seeker(pack_1, pack_2):
    for item in pack_1:
        if item in pack_2:
            return item
```

```python
example_bags = [
    'vJrwpWtwJgWrhcsFMMfFFhFp',
    'jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL',
    'PmmdzqPrVvPwwTWBwg',
    'wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn',
    'ttgJtRGJQctTZtZT',
    'CrZsJsPPZsGzwwsLwLmpwMDw'
]
```

```python
# bag check loop
values = []
for bag in bags:
    half = int(len(bag) / 2)
    #print(bag[:half], bag[half:])
    
    found = seeker(bag[:half], bag[half:])
    
    #print(priomap[found])
    values.append(priomap[found])
```

```python
sum(values)
```

```python
def seeker2(pack_1, pack_2, pack_3):
    for item in pack_1:
        if item in pack_2 and item in pack_3:
            return item
```

```python
badges = []
for i in range(0, 300, 3):
    badge = seeker2(*bags[i: i+3])
    badges.append(priomap[badge])
```

```python
sum(badges)
```
