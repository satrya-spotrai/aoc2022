```python
from pathlib import Path
import os
import shutil
```

```python
def file_writer(base, file):
    filesize, name = file.split(' ')
    name = '-'.join(name.split('.'))

    with open (base / f'{name}.txt', 'w') as handle:
        handle.write(filesize)
```

```python
def dir_writer(base, folder):
    os.mkdir(base / folder)
```

```python
current_path = Path('./data/root')
```

```python
for command in rcommands[1:]:

    if command[0] == '$':
        parts = command.split(' ')
        
        if parts[1] == 'cd':
            folder = parts[2]
            
            # move up
            if folder == '..':
                current_path = Path(os.path.join(*current_path.parts[:-1]))
            else:
                current_path = current_path / folder
        # # ls
        # else:
    
    # creat files/dirs
    else:
        if command.split(' ')[0] == 'dir':
            dir_writer(current_path, command.split(' ')[1])
        # create file
        else:
            file_writer(current_path, command)
```
