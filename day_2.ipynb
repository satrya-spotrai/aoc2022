```python
guide = {
    'Y': 'B',
    'X': 'A',
    'Z': 'C'
}
```

```python
winvals = {
    'A': 1,
    'B': 2,
    'C': 3
}
```

```python
def process(a, b):
    if a == b:
        return 3 + winvals[b]
    
    # rock
    if a == 'A':
        if b == 'B':
            return 6 + winvals['B']
        else:  # C
            return 0 + winvals['C']
    
    # paper
    if a == 'B':
        if b == 'C':
            return 6 + winvals['C']
        else:  # A
            return 0 + winvals['A']
    
    # scissor
    if a == 'C':
        if b == 'A':
            return 6 + winvals['A']
        else:  # B
            return 0 + winvals['B']
```
