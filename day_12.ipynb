```python
def convert_heightmap():
    height_map = []
    for line in lines:
        val_line = []
        for letter in line:
            if letter == 'S':
                val_line.append(-1)
            elif letter == 'E':
                val_line.append(27)
            else:
                val_line.append(letter_map[letter])
        
        height_map.append(val_line)
    
    return np.array(height_map)
```
